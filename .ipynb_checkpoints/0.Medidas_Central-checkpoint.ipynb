```python
# Importamos los modulos basicos para trabajar  con los datos
import numpy as np
from scipy import stats
import pandas as pd
```

- Modulo numpy: Es una libreria especializada en el calculo numerico y el analisis de datos, especialmente para un gran volumen de datos.

- Modulo stats: Es un modulo que ofrece funciones para el calculo de valores estadisticos en el campo de los numeros reales. Algunas de las fuciones que ofrece son:
  - mean(), devuelve la media aritmetica de los datos
  - median(), devuelve la mediana de los datos
  - mode(), devuelve la moda de los datos
  - stdev(), devuelve la desviacion estandar de la muestra 
  - variance() devuelve la varianza de la muestra de la poblacion
  - ...

- Modulo pandas: Es un modulo de Python que proporciona estructuras de datos similares a los dataframes de R. Pandas depende de Numpy y añade un potente tipo matricial a Python. Es una herramienta demasiado util para el analisis de datos.

#Lectura de los datos

```python
# Leemos el archivo utilizando la libreria pandas
# https://www.openintro.org/data/index.php?data=thermometry
data = pd.read_csv("/content/Temp_Genero_Ok.csv", header=0)

# Creamos un dataframe a partir de los generos
data_mujeres = data[data['gender'] == "female"]
data_hombres = data[data['gender'] == "male"]

# Eliminamos nuestra columna genero
data_mujeres = data_mujeres.drop(['gender'], axis =1)
data_hombres = data_hombres.drop(['gender'], axis =1)

# Organizamos los nombres de nuestras columnas
data_mujeres = data_mujeres.rename(columns={'body.temp':'Temperatura(F)','heart.rate':'Ritmo_cardiaco(L*s)'})
data_hombres = data_hombres.rename(columns={'body.temp':'Temperatura(F)','heart.rate':'Ritmo_cardiaco(L*s)'})
```

```python
# Buscamos una funcion para pasar de Cgrafos Fahrenheit a Celsius
def fahrenheit_a_celsius(f):
  return (f - 32) / 1.8, 2
  #return round((f - 32) / 1.8, 2)
```

```python
# Aplicamos la ecuacion a los dos dataframes que tenemos
data_mujeres['Temperatura(F)'] = data_mujeres['Temperatura(F)'].apply(fahrenheit_a_celsius)
data_hombres['Temperatura(F)'] = data_hombres['Temperatura(F)'].apply(fahrenheit_a_celsius)

# Renombramos nuevamente las columnas modificadas
data_mujeres = data_mujeres.rename(columns={'Temperatura(F)':'Temperatura(C)'})
data_hombres = data_hombres.rename(columns={'Temperatura(F)':'Temperatura(C)'})

# Reiniciamos el index
data_mujeres = data_mujeres.reset_index(drop=True)
data_hombres = data_hombres.reset_index(drop=True)
```

# Grafiquemos los datos

```python
import matplotlib.pyplot as plt

# Realicemos un histograma
data_hombres['Temperatura(C)'].hist(bins=8) 
plt.xlabel("Temperatura (°C)")
plt.ylabel("Frecuencia")
plt.show()
```

Como se puede ver en el dato, la temperatura mas frecuente en los hombres esta por el orden de los 36.75 °C. Gracias a este tipo de grafico podemos obtener una "primera vista" generl de la distribucuon de la poblacion o de la muestra respecto a una caracteristica.

```python
# Realicemos la distribucion de frecuencias
tabla_frec = pd.value_counts(data_hombres['Temperatura(C)'])
tabla_frec
```

```python
# Sacamos el maximo y el minimo para evaluar nuestro rango
Max = data_hombres['Temperatura(C)'].max()
Min = data_hombres['Temperatura(C)'].min()

print("Min:", Min," - Max:", Max)
```

```python
# Estimamos los pasos
# Formula : k = 1 + 3,322 * log(Tam)
K = 7.0224

# Calculamos el rango
Rango =  Max - Min

# Amplitud del intervalo
Ampl = Rango / K
print(Ampl)
```

```python
# Definimos los contenedores para la tabla de frecuencias
contenedores = np.arange(Min, Max, Ampl)

# Unimos las columnas
frec = pd.cut(data_hombres['Temperatura(C)'], contenedores)

# Sacamos la tabla de frecuencias
tabla_frec = pd.value_counts(frec)
print(tabla_frec)
```

Con base a la distribucion de frecuencia podemos decir que el rango de 36.482-36.735 y el rango 36.988-37.241 son aquellos que poseen mayor cantida de datos, siendo de 15 para ambos. Por lo que en esta muestra podemos concluir que las temperaturas mas comunes en los hombres esta entre esos dos rangos
