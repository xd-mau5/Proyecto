```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
data = pd.read_csv('https://files.xd-mau5.xyz/Datos/ICFES.csv', header=0)
data.head()
```

# Poblacion y Muestra
En este caso la poblacion seran las 1000 personas que presentaron las pruebas ICFES y nuestra muestra sera el puntaje global

```python
data['PUNT_GLOBAL'].hist(bins=5)
plt.title('Histograma de puntaje global')
plt.xlabel("Puntaje")
plt.ylabel("Frecuencia")
plt.show
```

Notamos que hay outliers en nuestros datos ya que sabemos que es imposible tener esos valores, organizaremos nuestros datos para encontrar los outliers

```python
data_sort = data.sort_values('PUNT_GLOBAL')
data_sort = data_sort.reset_index(drop=True)
data_sort['PUNT_GLOBAL'].tail(20)
```

Usando tail encontramos valores exagerados mostrados en la parte de arriba por lo tanto se descartaran para evitar problemas con estos


```python
data_sort = data_sort.iloc[:-12]
data_sort['PUNT_GLOBAL'].tail(20)
```

```python
#Graficamos nuevamente pero con los nuevos datos
data_sort['PUNT_GLOBAL'].hist(bins=10)
plt.title("Histograma de puntaje global")
plt.xlabel("Puntaje")
plt.ylabel("Frecuencia")
plt.show()
```

# Tabla de distribucion de frecuencias
Ahora que tenemos nuestros datos organizados podemos crear una tabla de distribucion de frecuencias para cada una de las columnas

```python
# Realicemos la distribucion de frecuencias
tabla_freq = pd.value_counts(data_sort['PUNT_GLOBAL'])
tabla_freq
```

```python
# Sacamos el maxio y el minimo para evaluar el rango
maximo = data_sort['PUNT_GLOBAL'].max()
minimo = data_sort['PUNT_GLOBAL'].min()
print("Min: ", minimo, "-Max: ", maximo)
# Calculamos el rango
rango = maximo - minimo
print("Rango: ", rango)
# # Estimamos los pasos
# Formula : k = 1 + 3,322 * log(Tam)
K = 1 + 3.3 * np.log10(len(data_sort['PUNT_GLOBAL']))
print("K: ", K)
# Calculamos la amplitud
amplitud = rango / K
print("Amplitud: ", amplitud)
# Calculamos los intervalos
intervalos = np.arange(minimo, maximo, amplitud)
# Unimos las columnas
freq = pd.cut(data_sort['PUNT_GLOBAL'], intervalos)
# Mostramos la tabla
tabla_freq = pd.value_counts(freq)
print("Tabla de frecuencias")
print(tabla_freq)
```

Con base a la tabla de distribucion de frecuencias podemos concluir que la mayoria de puntajes esta en el rango de 242.74 a 278.031, teniendo la mayor cantidad de datos con un numero de 268

Tenemos que la media de los puntajes globales es de 240.14

```python
#Calculamos la mediana
data_sort['PUNT_GLOBAL'].median()
```

Tenemos que la mediana de los puntajes globales es de 237

# ¿Qué genero presenta mejores puntajes en matematicas?

```python
# ¿Qué genero presenta mejores puntajes en matematicas?
# 1. Agrupamos por genero y puntaje de matematicas
data_sort.groupby(['GENERO'])['PUNT_MATEMATICAS'].mean()
# 2. Graficamos con grafico de barras
data_sort.groupby(['GENERO'])['PUNT_MATEMATICAS'].mean().plot(kind='bar', color=['red', 'blue'])
plt.title("Puntaje de matematicas por genero")
plt.xlabel("Genero")
plt.ylabel("Puntaje")
plt.show()
# 3. Conclusion: Falso, el genero masculino presenta mejores puntajes en matematicas
```

Viendo los resultados de la tabla de distribucion de frecuencias podemos concluir que el genero masculino obtuvo mejores puntajes en matematicas

# ¿Los colegios con ubicacion rural obtuvieron un mejor puntaje global?

```python
# ¿Los colegios con ubicacion rural obtuvieron un mejor puntaje global?
# 1. Agrupamos por ubicacion y puntaje global
data_sort.groupby(['AREA_UBICACION_COLEGIO'])['PUNT_GLOBAL'].mean()
# 2. Graficamos con grafico de barras
data_sort.groupby(['AREA_UBICACION_COLEGIO'])['PUNT_GLOBAL'].mean().plot(kind='bar', color=['red', 'blue'])
plt.title("Puntaje global por ubicacion")
plt.xlabel("Ubicacion")
plt.ylabel("Puntaje")
plt.show()
# 3. Conclusion: Falso, los colegios con ubicacion urbana obtuvieron un mejor puntaje global
```

Viendo los resultados de la tabla de distribucion de frecuencias podemos concluir que los colegios con ubicacion urbana obtuvieron un mejor puntaje global

# ¿Cuantas personas de estas 1000 tienen y no tienen internet en su casa?

```python
# ¿Cuantas personas de estas 1000 tienen y no tienen internet en su casa?
# 1. Agrupamos por internet y contamos
data_sort.groupby(['TIENEINTERNET'])['TIENEINTERNET'].count()
# 2. Graficamos con grafico de pastel
data_sort.groupby(['TIENEINTERNET'])['TIENEINTERNET'].count().plot(kind='pie', autopct='%1.1f%%')
plt.title("Personas con y sin internet")
plt.xlabel("Acceso a internet")
plt.ylabel("Personas")
plt.show()
# 3. Conclusion: 62.3% de las personas de estas 1000 tienen internet en su casa y 37.7% no tienen internet en su casa
```

Con base a la grafica anterior podemos concluir que el 62.3% de las personas tienen internet en su casa y el 37.7% no lo tiene

# ¿Cual de las 5 areas del conocimiento obtuvo menor y mayor puntaje en general?

```python
# ¿Cual de las 5 areas del conocimiento obtuvo menor y mayor puntaje en general?
# 1. Calculamos el puntaje promedio de cada area excepto la global
puntaje = data_sort.mean(numeric_only=True)[0:5]
# Cambiamos el nombre de las columnas
puntaje.index = ['Lectura Critica', 'Matematicas', 'Ciencias Naturales', 'Ciencias Sociales', 'Ingles']
print("Puntaje promedio por area:")
print(puntaje.to_string())
# 2. Graficamos con grafico de barras
puntaje.plot(kind='bar', color=['red', 'blue', 'green', 'yellow', 'purple'])
plt.title("Puntaje promedio por area")
plt.xlabel("Area")
plt.ylabel("Puntaje")
plt.show()
# 3. Imprimimos el area con menor y mayor puntaje
print("Area con menor puntaje: ", puntaje.idxmin(), " - Puntaje: ", puntaje.min())
print("Area con mayor puntaje: ", puntaje.idxmax(), " - Puntaje: ", puntaje.max())
# 4. Conclusion: La area de lectura critica obtuvo el mayor puntaje y la de ciencias sociales el menor
```

Con base a la grafica anterior podemos concluir que la area de lectura critica obtuvo el mayor puntaje y la area de ciencias sociales obtuvo el menor puntaje
