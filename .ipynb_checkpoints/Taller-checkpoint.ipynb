### La "ciencia de los datos" es un término erróneo
Lo primero es lo primero: *la inmensa mayoría de los trabajos que aprovechan las habilidades de datos NO son trabajos de “científico de datos”*. Esto significa que el término “ciencia de datos” es un
término erróneo. Familias de trabajos como “Analista de Datos”, “Ingeniero de Datos” y “Analista
de Inteligencia de Negocios” (entre otros) tienen un orden de magnitud más de ofertas de trabajo
que “Científico de Datos”, mientras que siguen aprovechando las habilidades de datos en un grado
considerable. De hecho, un gran número de empleos de “científico de datos” están mal clasificados
y encajarían mejor en una de las otras tres familias de empleos mencionadas anteriormente.

En cualquier caso, todos estos trabajos se engloban bajo el paraguas general de **profesionales de los datos**. Un profesional de los datos es alguien que combina los conocimientos de tres disciplinas
principales en alguna combinación (la combinación exacta dependerá de la familia de empleos):

**Matemáticas y estadísticas (principalmente estadísticas)**. A partir de aquí necesitamos las habilidades de análisis y modelización, que es el arte de deducir y representar cosas sobre el mundo utilizando principios lógicos. Los conocimientos matemáticos variados son a veces necesarios o muy útiles aquí
*Desde aquí necesitamos la habilidad de codificación o programación, que es la práctica de escribir instrucciones que un ordenador puede entender para dar vida a nuestro análisis y modelado.* Necesitamos esto para guiar nuestro análisis y modelado y para actuar correctamente
sobre los resultados.

### El lenguaje de programación
Para codificar algo para que nuestro ordenador lo haga, debemos comunicarnos con él en un lenguaje que tanto nosotros como el ordenador podamos entender. Este lenguaje común se conoce como **lenguaje de programación**. Un lenguaje de programación traduce instrucciones legibles por un humano en instrucciones legibles por el ordenador, lo que se conoce comúnmente como **código máquina**.

Los lenguajes de programación tienen algunas cosas en común con idiomas como el inglés o el español. Por ejemplo (¡y esto es muy importante!) son (o deberían ser) comprensibles para los humanos. Al igual que el inglés o el español, tienen reglas sobre cómo deben usarse (es decir, ortografía, gramática, sintaxis).

Sin embargo, los lenguajes de programación son mucho más estrictos a la hora de aplicar estas reglas que los lenguajes humanos. Si no se siguen exactamente las reglas de un lenguaje de programación,
tu programa no **compilará**. lo que  significa que el ordenador no ha sido capaz de traducir tus instrucciones a código máquina. Por otro lado, el hecho de que sean estrictas facilita que un grupo de personas lea y trabaje en el código de forma conjunta (una gran ventaja, ya que las grandes cosas suelen crearse en equipo), y hace que lo que hace el código sea inequívoco.

Hay muchos, muchos lenguajes de programación, pero para este programa y para la mayoría de los trabajos de datos que hay, hay un rey por encima de todos…

### Python
El lenguaje de programación de elección para el trabajo de datos hoy en día es **Python**. (Hay otros como R, Scala o Julia, pero están muy, muy por detrás en cuota de mercado y en la cantidad de funcionalidad que ofrecen para el trabajo con datos).

![python](Imagenes/python.png)

Python es genial por 3 razones importantes:

- Es fácil de aprender
- Es fácil de leer (¡muchos otros lenguajes de programacion no lo son!)
- Es de código abierto. lo que significa que es gratuito y, lo que es mejor, que otras personas de la comunidad de Python añaden constantemente más funcionalidad al lenguaje de forma gratuita.

Cuando vas a codificar un programa, casi nunca empiezas desde cero normalmente usas código que fue construido por alguien antes que hace una determinada cosa. Por ejemplo, si quiero escribir un programa que analice datos y necesite encontrar la media de un conjunto de números, ¡no necesito escribir yo mismo el código que encuentra la media! Dado que se trata de un problema común, es probable que alguien ya haya creado código específicamente para ese propósito. Entonces puedo descargar esa funcionalidad de Internet importarla a mi programa y utilizarla sin muchas complicaciones.

Estas funcionalidades escritas por otras personas y utilizables por cualquiera en un lenguaje se llaman **bibliotecas** o **paquetes**. Python tiene bibliotecas para cualquier cosa que se te ocurra. Para el ejemplo anterior, está la biblioteca `numpy`, que contiene funcionalidad para numerosos cálculos matemáticos. ¿Quieres algo para interactuar con archivos de Excel? Puedes usar `openpyxl`.
¿Imágenes? Tienes `Pillow`. Hay algo para todo: en este momento hay más de 200.000 paquetes de Python disponibles públicamente.

### El IDE
Habrás notado que el archivo que estás leyendo vive dentro de un “contenedor” llamado **Jupyter**. ¿Cuál es el problema con Jupyter?

Para entenderlo, primero debemos entender qué es un **Entorno de Desarrollo Integrado (IDE)**. Un IDE es la principal herramienta que vas a ver cuando escribas código. Los archivos de código no son en realidad más que archivos de texto, y podrían escribirse igualmente en Microsoft Word, el Bloc de notas o cualquier otro editor de texto. Pero ninguno de esos editores de texto básicos proporciona otras funcionalidades de codificación esenciales, como comprobar si has escrito una palabra clave en un lugar equivocado, o permitirte ejecutar tu código inmediatamente para probar si funciona.

Esto significa que los IDEs son esencialmente editores de texto con esteroides, con funcionalidades que ayudan mucho al proceso de codificación. Uno de los más populares para el trabajo con Python y datos es Jupyter porque hace algo que casi ningún otro IDE permite: ¡le permite escribir y ejecutar código junto con el texto estándar (similar a la combinación de un documento de Word
con un archivo de código)!

### Jupyter
Esta capacidad es lo que hace que Jupyter sea muy especial para la creación de prototipos de proyectos, la prueba de soluciones y la creación de entregables que incluyan no sólo código, sino también gráficos y texto y explicaciones. ¿Recuerdas cuando dijimos que realizar análisis de datos requería no sólo codificar, sino también aprovechar el conocimiento del dominio? Una de las partes, si no la más importante, del análisis de datos es explicar los resultados a los demás, y Jupyter lo hace muy fácil.

## Celdas
Verá una pantalla casi vacía, con una línea que dice `En [ ]:.` Esta línea se conoce como celda de código:

![empty](Imagenes/empty.png)

Los cuadernos Jupyter tienen un bloque de construcción llamado celda. Las celdas pueden ser de varios tipos, pero las dos más importantes son las **celdas de código** y las **celdas de Markdown**. Las celdas de código, como la que tienes en tu archivo vacío, te permiten escribir y ejecutar código en el cuaderno. Las celdas Markdown, como la que estás leyendo ahora, te permiten colocar texto con formato (y otros elementos que no son de código, como la imagen de arriba) en el cuaderno.

A continuación encontrarás una celda de código que calcula la suma de dos números e imprime el resultado en la pantalla. Mientras que las celdas de Markdown sólo tienen texto, las celdas de código tienen dos partes. La primera parte muestra el código en sí, y la segunda parte muestra lo que el código imprime después de ejecutarlo (haz clic en el bloque de código de abajo y luego pulsa `Ctrl+Enter` en tu teclado):

```python
result = 3 + 4
print(f"El resultado de 3 mas 4 es {result}")
```

Una cosa interesante de las celdas de código es que puedes editarlas y ejecutarlas de nuevo sin problemas. Por ejemplo, intente cambiar el valor 3 a 5 en la celda de código de arriba, y luego presione Ctrl+Enter para ejecutarla. El texto de abajo debería cambiar en consecuencia.

### Edición y creación de celdas
Puede convertir fácilmente una celda de código en una celda de Markdown, y viceversa. El párrafo de la siguiente imagen contiene algo de código, pero está en una celda Markdown, por lo que el código no se ejecuta. Para cambiarlo a una celda de código, sólo tienes que hacer doble clic en ella y en el menú desplegable de la parte superior, cambiar la opción de Markdown a Código. Luego presiona `Ctrl+Enter` para ejecutarlo.

![markdown](Imagenes/changeCell.jpg)

```python
text1 = 'Esto es una celda de código'
text2 = ' y Jupyter Notebooks son increibles!'
finalText = text1+text2
print(finalText)
```

Ahora debería poder ver el texto debajo de la celda de código mostrando una frase completa. Puede seguir adelante y editar los componentes de la frase para hacer que el código imprima algo diferente.

También puede crear y eliminar celdas - en la barra de herramientas de la parte superior izquierda, hay un signo “+” y un signo de tijera. El signo “+” crea una nueva celda inmediatamente debajo de la celda que tiene seleccionada, y el signo de las tijeras borra la celda que tiene seleccionada.

Hay muchos otros botones útiles en la barra de herramientas superior, como copiar, pegar o reordenar celdas. Todas estas acciones tienen su correspondiente atajo de teclado. El atajo más importante es `Ctrl+S`, que guarda tu trabajo. Asegúrate de usarlo a menudo. Para ver una lista completa de atajos, ve a [aquí](https://towardsdatascience.com/jypyter-notebook-shortcuts-bf0101a98330).


Algunos de los atajos más útiles son `* b -` crear una nueva celda debajo de la celda actual `* a -` crea una nueva celda por encima de la celda actual `* c”` - copiar la celda actual `* p-` pegar una celda copiada `*dd-` borrar la celda actual `*m‘ -` convertir la celda actual en una celda de marcado `* Y -` convertir la celda actual en una celda de código

Nota: Es necesario pulsar la tecla de escape antes de utilizar estos atajos.

Para ver todas las opciones de formato que tienes a tu disposición, consulta esta [hoja de trucos](https://www.markdownguide.org/cheat-sheet/).

### Controlar la ejecución de las celdas de código
Hay otros cuatro botones en el menú principal que queremos destacar. “Ejecutar” ejecuta el código en la celda actual, lo cual es bastante sencillo (`Ctrl+Enter` también ejecuta la celda actual).

El botón cuadrado negro es una señal de “Stop” e interrumpe el **núcleo** (piensa en el núcleo como el “motor” que ejecuta tu código en el entorno Jupyter). Esto significa que interrumpe cualquier ejecución de código que pueda estar ocurriendo en ese momento. Es importante saber que este botón existe, porque a veces puedes escribir un programa que se atasca y tarda mucho más de lo esperado o que claramente está haciendo algo totalmente diferente a lo que esperabas. En ese caso, puedes detener tu código prematuramente con este botón.

El botón de la flecha circular reinicia el núcleo. Esta es una versión más extrema de detener el núcleo. Cuando ejecutas código, el núcleo guarda tus resultados y variables mientras trabajas.
Cuando lo detienes, esas variables siguen ahí. Pero cuando se reinicia, todas las variables se pierden y se vuelve a una pizarra en blanco. A veces, detener el kernel no es suficiente para resolver ciertos problemas, por lo que es necesario reiniciarlo por completo.

El último botón con las 2 flechas hacia delante ejecuta todas las celdas del cuaderno de forma secuencial. Después de algún tiempo, habrá muchas celdas en su cuaderno, y usted no querrá pasar y ejecutar cada una manualmente. Este botón le permite ejecutar todo en un solo clic.

## El terminal
La última herramienta básica del oficio es el **Terminal** de tu ordenador. Este es un lugar que asusta a algunos, pero no te preocupes. Con el tiempo verás que la Terminal no sólo es tu amiga sino tu mayor aliada para hacer que las cosas sucedan rápidamente.

Una Terminal es lo que tenían los ordenadores antes de que Microsoft y Apple popularizaran las **Interfaces Gráficas de Usuario (GUI)** y los ratones de ordenador. Por aquel entonces, se escribían los comandos directamente en el ordenador en lugar de hacer clic en los botones. Así, si querías abrir el editor de MS-DOS (el precursor de Microsoft Word), escribías “editar” y pulsabas `Enter`. Tiempos oscuros, sin duda.

Para abrir tu Terminal, búscalo en el menú de tu máquina o escríbelo en su barra de búsqueda. Una vez que tengas tu Terminal abierto, verás una ventana con algo de texto y un cursor parpadeante. Aquí es donde se escriben los comandos a su ordenador. El aspecto es un poco diferente para cada sistema operativo - aquí hay algunos ejemplos para Windows, macOS y Linux (en orden):

### Propiedades y funcionalidades básicas
El mundo en tu Terminal es un poco diferente al de fuera de él. ¡La diferencia más importante es que `Ctrl+C` y `Ctrl+V` ya no copian y pegan! Dentro de tu Terminal, `Ctrl+C` realmente (C)anula el programa que se está ejecutando. Si quieres copiar o pegar información desde la Terminal de Jupyter, tienes que usar `Ctrl+Insert` (o `Cmd+C` para macOS) para copiar y `Shift+Insert` (o `Cmd+V` para macOS) para pegar. En caso de duda, siempre puedes hacer clic con el botón derecho para que aparezcan las opciones de copiar y pegar.
También hay algo de terminología que es importante: (1), cuando algo aparece en la terminal, decimos que se *imprime* en la terminal; (2) lo que normalmente llamamos carpetas (donde se agrupan los archivos) se llama ahora *directorios*; (3) tu directorio de trabajo actual es el directorio en el que te encuentras actualmente; y (4) si escribes un comando en la terminal y luego escribes algo después de ese comando, ese algo después se llama los *parámetros* del comando.
